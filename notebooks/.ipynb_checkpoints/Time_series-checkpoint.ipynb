 # 1. Price Trend Metrics

```python
# a. trends over time  
plt.title('Close over time')
plt.xlabel('time')
plt.ylabel('close')
plt.plot(raw_data['close'],label='close',color='green')
plt.figure(figsize=(10,8))
plt.show()
```

```python
raw_data['daily_returns']=raw_data['close'].pct_change()
raw_data['daily_returns']
```

```python
# c  Volatility = standard deviation

raw_data['volatility']=raw_data['daily_returns'].std()
raw_data['volatility']
```

```python
# high_volatile->high risk to predict
# low_volatile->stable ,consistent
# rolling volatility ->This helps see how risk changes over time.

```

```python
# Moving Averages
raw_data['rolling_volatility']=raw_data['daily_returns'].rolling(window=20).std()
raw_data['rolling_volatility']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(raw_data['time'], raw_data['rolling_volatility'], color='purple', label='30-period Rolling Volatility')

plt.title('Price Stability')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Boxplot for each numerical column
numeric_cols = raw_data.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=raw_data[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()
```

```python
# Detect outliers using IQR method
Q1 = raw_data[numeric_cols].quantile(0.25)
Q3 = raw_data[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Filtering the data
outliers = ((raw_data[numeric_cols] < (Q1 - 1.5 * IQR)) | (raw_data[numeric_cols] > (Q3 + 1.5 * IQR)))

print("Outlier Summary:")
print(outliers.sum())  # Shows how many outliers in each column
```

```python
# removing of outliers
from scipy.stats import zscore
raw_data['z_score'] = zscore(raw_data['close'])
raw_data[raw_data['z_score'].abs() > 3]  # These are outliers
```

```python
# 
#6.Outliers 
# Boxplot for each numerical column

numerical_cols=raw_data.columns
# print(raw_data.columns)

# removing of outliers(IQR Method)

#     Q1=df['col'].quantile(0.25)
#     Q3=df['col'].quantile(0.75)
    
#     IQR=Q3-Q1

#     # define Bounds
#     lower_bound=Q1-1.5*IQR
#     upper_bound=Q3+1.5*IQR

# df_no_outliers=df[(df['col']>=lower_bound)& (df['col']<=upper_bound)]

def remove_outliers(data,column):
    Q1=data[column].quantile(0.25)
    Q3=data[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q1+1.5*IQR
    return data[(data[column]>=lower_bound)&(data[column]>=upper_bound)]
    
for col in numerical_cols:
    before=raw_data.shape[0]
    no_outliers_data=remove_outliers(raw_data,col)
    after=no_outliers_data.shape[0]
    print(f'{col}: \n removed  {before-after}  outliers')
no_outliers_data.to_csv(r'C:\Users\Hi\Finlang\data\processed\no_outliers_data.csv')
print('no_outliers_data saved successfully')

    
# verification for outliers
numeric_cols=no_outliers_data.columns
for col in numeric_cols:
    plt.figure(figsize=(7,7))
    sns.boxplot(data=raw_data[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()
```
