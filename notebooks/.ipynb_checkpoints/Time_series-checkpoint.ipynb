 # 1. Price Trend Metrics

```python
# a. trends over time  
plt.title('Close over time')
plt.xlabel('time')
plt.ylabel('close')
plt.plot(raw_data['close'],label='close',color='green')
plt.figure(figsize=(10,8))
plt.show()
```

```python
raw_data['daily_returns']=raw_data['close'].pct_change()
raw_data['daily_returns']
```

```python
# c  Volatility = standard deviation

raw_data['volatility']=raw_data['daily_returns'].std()
raw_data['volatility']
```

```python
# high_volatile->high risk to predict
# low_volatile->stable ,consistent
# rolling volatility ->This helps see how risk changes over time.

```

```python
# Moving Averages
raw_data['rolling_volatility']=raw_data['daily_returns'].rolling(window=20).std()
raw_data['rolling_volatility']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(raw_data['time'], raw_data['rolling_volatility'], color='purple', label='30-period Rolling Volatility')

plt.title('Price Stability')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
